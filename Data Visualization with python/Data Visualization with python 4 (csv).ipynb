```python
import pandas as pd
```

```python
#read the data + setting the index column to the date + parsing the dates(storing them as actuall dates not just strings)
df = pd.read_csv('Fremont_Bridge_Bicycle_Counter.csv',index_col='Date', parse_dates=True)
```
